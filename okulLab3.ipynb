```python
dizi = []
dizi.append(int(input("1. sayıyı giriniz: ")))
for i in range(9):
    print(i,end = " ")
    sayi = int(input("Sayıyı giriniz: "))
    
    
    while sayi <= dizi[i]:
        print("Girdiğiniz sayı büyük olmalı")
        print(i, end=" ")
        sayi = int(input("Sayıyı girin: "))
    dizi.append(sayi)
print("\n")

for i in range(10):
    print(dizi[i],end=" ")
print("\n")
for i in range(10):
    print(dizi[9 - i],end = " ")
```

```python
from random import randint
list = []
sayac = 0
for i in range(20):
    list.append(int(randint(0,25)))

for i in range(20):
    sayac += list[i]
    if sayac > 100:
        break;
    print(list[i])
    
print(list)
```
